```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3]
y = [4,5,6]
plt.title("our first graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[4,5,6 ]
plt.title("our second graph", {'fontsize':25, 'color':'red'})
plt.xlabel("x-axis", {'fontsize':25, 'color':'blue'})
plt.ylabel("y-axis",{'fontsize':25, 'color':'orange'})
plt.plot(x,y, color='green')
plt.show()
```

```python
x=[1,2,3]
y=[4,5,6 ]
plt.title("our second graph", fontdict= {'fontsize':25, 'color':'red'})
plt.xlabel("x-axis", fontdict= {'fontsize':25, 'color':'blue'})
plt.ylabel("y-axis",fontdict = {'fontsize':25, 'color':'orange'})
plt.plot(x,y, color='green')
plt.show()
```

```python
dataFrame = pd.read_csv("/content/petrol_prices.csv")
dataFrame
```

```python
for i in dataFrame.columns[1:]:
  plt.plot(dataFrame['Year'], dataFrame[i], label=i)
plt.legend()
plt.show()
```

```python
plt.plot(dataFrame['Year'], dataFrame['Canada'],marker='*')
plt.plot(dataFrame['Year'], dataFrame['USA'],marker='o',label="USA")
plt.plot(dataFrame['Year'], dataFrame['Canada'],marker='o',label="Canada")
plt.legend()


```

```python
plt.title("Petrol prices over time in different countries", fontdict= {'fontsize':20, 'color':'#d08d88', 'fontname':'sans-serif'})
plt.xlabel("Years", fontdict= {'fontsize':25, 'color':'blue'})
plt.ylabel("Price in dollars",fontdict = {'fontsize':25, 'color':'orange'})

plt.plot(dataFrame['Year'], dataFrame['USA'],marker='o',label="USA")
plt.plot(dataFrame['Year'], dataFrame['Japan'],marker='o',label="Japan")
plt.plot(dataFrame['Year'], dataFrame['Australia'],marker='o',label="Australia")
plt.legend()
```
